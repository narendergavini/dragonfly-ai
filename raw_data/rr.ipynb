```python
import json
import os

# Read the JSON file
with open('cleaned_knowledge_base.json', 'r') as file:
    data = json.load(file)

# Create cleaned_data directory if it doesn't exist
os.makedirs('cleaned_data', exist_ok=True)

# Extract each key-value pair and save as text files
for key, value in data.items():
    filename = f"cleaned_data/{key}.txt"
    
    # Convert list to string with newlines if value is a list
    if isinstance(value, list):
        content = '\n'.join(str(item) for item in value)
    else:
        content = str(value)
    
    # Save to file
    with open(filename, 'w', encoding='utf-8') as txt_file:
        txt_file.write(content)
    
    print(f"Saved {key} to {filename}")

print(f"\nTotal files created: {len(data)} in cleaned_data folder")
```

```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

# 1. Load .doc file
loader = UnstructuredWordDocumentLoader("Procrastination DF SCRIPTS.doc")
docs = loader.load()

# 2. Split documents into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=150
)
splits = text_splitter.split_documents(docs)

# 3. Create embeddings and store
embeddings = OpenAIEmbeddings()
vector_store = FAISS.from_documents(splits, embeddings)

# 4. Save locally
vector_store.save_local("./vector_store")

# 5. Load later
vector_store = FAISS.load_local("./vector_store", embeddings, allow_dangerous_deserialization=True)

```

```python
# 6. Retrieve and generate
retriever = vector_store.as_retriever(search_kwargs={"k": 2})
query = "What are some strategies to overcome procrastination?"
retrieved_docs = retriever.invoke(query)
```

```python
retrieved_docs
```

```python
# 7. Pass to LLM for answer generation
llm = ChatOpenAI(model="gpt-4o-mini")
context = "\n\n".join([doc.page_content for doc in retrieved_docs])
response = llm.invoke(f"Context: {context}\n\nQuestion: {query}")
print(response.content)
```
